### 2개의 축을 갖는 그래프 작성
- 남북한 발전량 데이터 셋을 사용
- 막대그래프는 북한의 발전량을 표시
- 선그래프는 북한 발전량의 전년대비 증감률을 백분율로 나타낸다

```python
df = pd.read_excel("../Data/남북한발전전력량.xlsx")
df
```

```python
# 북한 전력량만 선택
df = df.loc[5:9]
df
```

```python
df.drop("전력량 (억㎾h)", axis = 1, inplace = True)
df
```

```python
# index를 발전 전력별로 변경
df.set_index("발전 전력별", inplace=True)
df
```

```python
df = df.T
df.head()
```

```python
# 합계 컬럼을 총 발전량으로 변경
df = df.rename(columns={"합계":"총발전량"})
df.head()
```

```python
# 증감율(변동율) 계산
# 총발전량 - 1년 이라는 컬럼 생성
df["총발전량 - 1년"] = df["총발전량"].shift(1)
df.head()
```

```python
# 증감률 컬럼 생성
df["증감율"] = ((df["총발전량"] / df["총발전량 - 1년"]) - 1) * 100
df.head()
```

```python
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
plt.style.use("ggplot")
# 2축 그래프 그리기

ax1 = df[["수력", "화력"]].plot(kind="bar", figsize=(20,10), stacked=True)

# 같은 x 축을 공유
ax2 = ax1.twinx()
ax2.plot(df.index, df.증감율, ls="--", marker="o", markersize=10, color="green")

ax1.set_ylim(0, 500)
ax2.set_ylim(-50, 50)

ax1.set_xlabel("연도", size=20)
ax1.set_ylabel("발전량(억 kWh)")
ax2.set_ylabel("전년대비 증감율(%)")

plt.title("북한 전력 발전량(1990~2016)", size = 30)
ax1.legend(loc="upper left", fontsize=20)
plt.show()
```

### 히스토그램(histogram)
: 단위별(컬럼별) 데이터의 빈도수를 그래프로 표현

```python
# Style 서식 지정
plt.style.use("classic")
```

```python
df = pd.read_csv("../Data/auto-mpg.csv", header=None)
df.head()
```

```python
df.columns = ['mpg','cylinders','displacement','horsepower','weight',
              'acceleration','model year','origin','name']
df.head()
```

```python
# 연비(mpg) 열에 대한 히스토그램 그리기
df["mpg"].plot(
                kind = "hist",
                bins = 10, # 몇개로 나눠서 그릴 것인지
                color = "coral",
                figsize = (10, 5)
)
# 그래프 그리기
plt.title("Historgram")
plt.xlabel("mpg")

plt.show()
```

> 자동차의 연비값은 대부분 좌측 편향이 되어 연비가 낮은 구간에 집중되어 있는 경향을 보인다.

### 산점도 (Scatter plot)
- 서로 다른 두 변수 사이의 관계를 나타낸다.

```python
plt.style.use("default")
```

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
  kind = "scatter",
  x = "weight",
  y = "mpg",
  c = "coral", # color로 바꿔도 됨
  s = 10, # 점 크기
  figsize = (10, 5)
)
plt.title("Scatter Plot = mpg vs. weight")
plt.show()
```

> 차중과 연비는 서로 강한 역 상관관계를 갖는다.

```python
# Cylinder 개수의 상대적 비율을 계산하여 시리즈 생성
# 그래프를 그리기 위해 크기를 늘리려 300(Hyper parameter)을 곱함
cylinder_size = df.cylinders / df.cylinders.max() * 300
print(cylinder_size)
```

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
  kind = "scatter",
  x = "weight",
  y = "mpg",
  c = "coral", # color로 바꿔도 됨
  s = cylinder_size, # 점 크기
  figsize = (10, 5)
)
plt.title("Scatter Plot = mpg vs. weight")
plt.show()
```

### 버블(Bubble) 차트

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
  kind = "scatter",
  x = "weight",
  y = "mpg",
  c = "blue", # color로 바꿔도 됨
  s = cylinder_size, # 점 크기
  alpha = 0.3, # 투명도
  figsize = (10, 5)
)
plt.title("Scatter Plot = mpg vs. weight")
plt.show()
```

```python
# 연비(mpg)와 차중(weight) 열에 대한 산점도 그리기
df.plot(
  kind = "scatter",
  x = "weight",
  y = "mpg",
  c = cylinder_size, # color로 바꿔도 됨
  s = 50, # 점 크기
  alpha = 0.3, # 투명도
  figsize = (10, 5),
  marker = "+",
  cmap = "viridis" # 오른쪽의 color bar로 color map 보여주기
)
plt.title("Scatter Plot = mpg vs. weight")
plt.show()
```

### 파이차트(pie chart)
- 원을 파이처럼 나누어서 표현
- 조각의 크기는 해당변수에 속하는 데이터 값의 크기에 비례

#### 자동차의 국가별 데이터 개수를 파이차트로 표현

```python
plt.style.use("default")
```

```python
df.head()
```

```python
# 국가별 Count를 계산
df["count"] = 1
df.head()
```

```python
# origin열을 기준으로 그룹화, 합계 연산
df_origin = df.groupby("origin").sum()
df_origin
```

```python
# index값을 변경
df_origin.index = ["USA", "EU", "JAPAN"]
df_origin
```

```python
# 파이차트 그리기
df_origin["count"].plot(
  kind = "pie",
  figsize = (7,5),
  autopct = "%.2f%%", # 자동으로 퍼센트 만들어서 보여주기 (%.2f --> float 소수점 2째자리까지), (%% --> % 글자 출력)
  startangle = 90, # 시작 각도
  colors = ["chocolate", "bisque", "cadetblue"]
)

plt.title("Model Origin", size = 20)
plt.axis("equal") # 가끔 파이차트가 찌그러질때가 있는데 그때를 위해서 쓰는 코드
plt.legend(labels=df_origin.index, loc = "best")
plt.ylabel("")

plt.show()
```

### 박스플롯(boxplot)
- 범주형 데이터의 분포를 파악

##### 제조 국가별 연비 분포 박스플롯

```python
plt.style.use("seaborn-poster")
```

```python
# 한글 폰트 문제 해결 
# matplotlib은 한글 폰트를 지원하지 않음
# os정보
import platform

# font_manager : 폰트 관리 모듈
# rc : 폰트 변경 모듈
from matplotlib import font_manager, rc
# unicode 설정
plt.rcParams['axes.unicode_minus'] = False

if platform.system() == 'Darwin':
    rc('font', family='AppleGothic') # os가 macos
elif platform.system() == 'Windows':
    path = 'c:/Windows/Fonts/malgun.ttf' # os가 windows
    font_name = font_manager.FontProperties(fname=path).get_name()
    rc('font', family=font_name)
else:
    print("Unknown System")
```

```python
# 그래프 객체 생성
fig = plt.figure(figsize=(15, 5))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)

# 수직 박스 플롯
ax1.boxplot(
  x = [
    df[df["origin"] == 1]["mpg"],
    df[df["origin"] == 2]["mpg"],
    df[df["origin"] == 3]["mpg"]
  ],
  labels = df_origin.index
)
ax2.boxplot(
  x = [
    df[df["origin"] == 1]["mpg"],
    df[df["origin"] == 2]["mpg"],
    df[df["origin"] == 3]["mpg"]
  ],
  labels = df_origin.index,
  vert = False, # 수평으로 바꿔주는 것
)



ax1.set_title("제조 국가별 연비 분포(수직 박스 플롯)",size = 15)
ax2.set_title("제조 국가별 연비 분포(수평 박스 플롯)",size = 15)
plt.show()
```
