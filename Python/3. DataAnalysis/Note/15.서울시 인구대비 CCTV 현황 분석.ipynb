### 인구대비 CCTV 비율

```python
data_result["인구대비CCTV"] = data_result.소계 / data_result.인구수 * 10000
data_result.head()
```

```python
data_result.인구대비CCTV.sort_values().plot(
              kind="barh", 
              color= "cornflowerblue",
              figsize=(10,10)
              )
plt.ylabel("구별")
plt.xlabel("소계")
plt.title("인구수 10000명당 CCTV 비율")
plt.show()
```

> 상위 5개지역은 서울 중심부에 위치, 하위 5개지역은 서울 외곽지역에 위치해 있다   
> 송파구의 경우 인구수는 제일 많고, CCTV 대수는 낮은 편으로 비율이 가장 낮다. 그리고 고령자 수는 제일 많지만 고령자 비율이 가장 낮은 지역이다.   
> 중구는 인구수가 제일 적고, CCTV 대수도 낮은편이므로 가장 비율이 높다   
> 강남구는 인구수도 3등으로 많고 CCTV 대수도 가장 많으므로 비율이 높다   

```python
# 산점도를 이용한 시각화 (인구수와 소계)
data_result.plot(kind = "scatter", x="인구수", y = "소계")
plt.ylabel("CCTV")
plt.show()
```

```python
# 인구수, 소계에 따른 기준직선(회귀선, 평균선) 만들기
fp1 = np.polyfit(data_result["인구수"], data_result["소계"], 1)
fp1
```

```python
# x축 만들기 : 인구수 표현
fx = np.linspace(100000, 700000)

# y축 데이터 만들기 : CCTV 소계 표현
f1 = np.poly1d(fp1)

plt.figure(figsize=(10,10))
plt.scatter(data_result.인구수, data_result.소계, color = "blue")
plt.plot(fx, f1(fx), ls = "dashed", lw=3, color = "g")
plt.xlabel('인구수')
plt.ylabel("CCTV")
plt.show()
```

> 이 직선은 CCTV 댓수에 기준값이 됩니다.   
> 이 직선을 기준으로 상위에 있는 점은 과하게 설치된 것이고    
> 직선 보다 하위에 있는 점은 인구수에 비해 CCTV가 부족하다는 해석이 됩니다.

```python
# 기준선 구하기
fp1 = np.polyfit(data_result.인구수, data_result.소계, 1)

# 기준치 구하기
f1 = np.poly1d(fp1)

# 인구수
fx = np.linspace(100000, 700000)

# 소계에서 기준인구수 차이로 오차값 구하기
data_result["오차"] = np.abs(data_result.소계 - f1(data_result.인구수))

# 정렬
df_sort = data_result.sort_values("오차", ascending=False)
df_sort.head()
```

```python
# 그래프에 텍스트와 color
plt.figure(figsize=(14,10))
plt.scatter(data_result.인구수, data_result.소계, c=data_result.오차)
plt.plot(fx, f1(fx), ls = "dashed", lw=3, color = "g")

# 오차가 가장 큰 10개 구에 텍스트 붙이기
# plt.text(500000, 3000, "Test")
for n in range(10):
  plt.text(df_sort.인구수[n] * 1.02, df_sort.소계[n] * 0.99, df_sort.index[n], fontsize=15)

plt.xlabel("인구수")
plt.ylabel("CCTV")
plt.colorbar()
plt.show()
```
