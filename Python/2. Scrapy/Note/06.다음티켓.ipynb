```python
# Library
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

# Site Address
url = "https://movie.daum.net/ranking/boxoffice/weekly"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup 으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
li_lists = soup.select("#mainContent > div > div.box_boxoffice > ol > li > div > div.thumb_cont > strong > a")
print(li_lists)
```

```python
count = 0
titleLists = [] # 영화 제목 List
countLists = [] # 순위 Lists
for list in li_lists:
  count += 1
  print(count, list.string.strip())
  titleLists.append(list.string.strip())
  countLists.append(count)
```

```python
# Lists 확인
print(countLists)
print(titleLists)
```

```python
# countLists만으로 DataFrame 생성
df = pd.DataFrame(data = countLists, columns= ["순위"])
df
```

```python
# 기존 DataFrame에 컬럼과 Lists 추가
df["영화제목"] = pd.Series(titleLists)
```

```python
df
```

```python
# csv로 저장

df.to_csv("../Data/DaumList.csv", encoding= "utf-8", index=False)
```

```python
# Library
from bs4 import BeautifulSoup
import urllib.request as req
import pandas as pd

# Site Address
url = "https://movie.naver.com/movie/sdb/rank/rmovie.naver"

# urlOpen
res = req.urlopen(url)

# BeautifulSoup 으로 분석하기
soup = BeautifulSoup(res, "html.parser")
print(soup)
```

```python
nmovie = soup.select("tbody a")
print(nmovie)
```

```python
count = 0
countLists = []
titleLists = []
for i in nmovie:
  count += 1
  print(count, i.string.strip())
  countLists.append(count)
  titleLists.append(i.string.strip())
```

```python
df = pd.DataFrame(data=countLists, columns=["순위"])
df["영화제목"] = pd.Series(titleLists)
df
```

```python
df.to_csv("../Data/naverRanking.csv" , encoding="utf-8", index=False)
```
